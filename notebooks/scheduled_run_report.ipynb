```python
import pandas as pd
from IPython.display import clear_output, Image
from scheduled_job import score_loans

cm = {-2500: '#db5f57',
-2000: '#db7157',
 -1500: '#db8457',
  -1000: '#db9657',
-500:  '#dba957',
0:  '#dbbb57',
500:  '#dbce57',
1000:  '#d6db57',
1500:  '#c4db57',
2000:  '#b1db57'}

def highlight(row):
    
    for thres, col in cm.items():
        if row['expected_profit'] < thres:
            return ['background-color: {0}'.format(col) for i in range(len(row))]
    
    return ['' for i in range(len(row))]

clear_output()
```
